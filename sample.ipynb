```python
import pandas as pf
a = pf.read_csv("data/report.csv")
def leep(Year):
    if Year%4==0:
        return "yes"
    else:
        return "no"
a["leep"] = a["Year" ].apply(leep)
a
a.to_csv("data/leep.csv",index = False)
```
